#### For language worked with and language want to work with I've created a dictionary that counts the people using the most popular programming languages and languages they want to work with.

```python
survey_df['LanguageHaveWorkedWith'] =survey_df['LanguageHaveWorkedWith'].apply(str)
```

```python
lang_list = ['Java','JavaScript','Python','SQL','C#','PHP','HTML/CSS','R','Swift','C++','Assembly']
```

```python
lang_dict_have = {}
for i in survey_df['LanguageHaveWorkedWith']:
    for j in lang_list:
        if j in i:
            lang_dict_have[j]=lang_dict_have.get(j,0)+1
```

```python
print(lang_dict_have)
```

```python
survey_df['LanguageWantToWorkWith'] =survey_df['LanguageWantToWorkWith'].apply(str)
```

```python
lang_dict_want = {}
for i in survey_df['LanguageWantToWorkWith']:
    for j in lang_list:
        if j in i:
            lang_dict_want[j]=lang_dict_want.get(j,0)+1
```

```python
print(lang_dict_want)
```

#### For Salary since the data is given in the form of total income of users in different periods, I've made a column to calculate the their income in terms of income/year.

```python
salary = survey_df[~survey_df.CompTotal.isnull()]
```

```python
salary['salary'] = 0
```

```python
salary.loc[salary.CompFreq == 'Yearly','salary'] = salary.CompTotal
salary.loc[salary.CompFreq == 'Monthly','salary'] = salary.CompTotal * 12
salary.loc[salary.CompFreq == 'Weekly','salary'] = salary.CompTotal * 52
survey_df.loc[survey_df.YearsCodePro == 'Less than 1 year','YearsCodePro'] = '0'
```

```python
survey_df['salary'] = salary['salary']
```

#### Since the income varies drastically over million and knowing that vast majority of people in this don't have income over a million per year, I've decided to seperate those data and work with income below million.

```python
salary_under_million = survey_df[survey_df.salary < 1000000]
```

```python
salary_over_million = survey_df[survey_df.salary > 1000000]
```

```python
salary_under_million.describe()
```

```python
sal_under_mil_freq_count = salary_under_million.groupby('Country')['Country'].transform('size') > 500
sal_under_mil_freq_count = salary_under_million[sal_under_mil_freq_count]
```

```python
countries_salary = sal_under_mil_freq_count.groupby('Country')[['salary']].mean()
```

```python
value_have = []
key_have = []
value_sum = 0
for k,v in lang_dict_have.items():
    value_have.append(v)
    key_have.append(k)
    value_sum = value_sum + v
```

```python
list(lang_dict_have.values())
```

```python
sns.barplot(x=list(lang_dict_have.values()),y=list(lang_dict_have.keys()))
plt.title("Number of respondents using the languages")
plt.xlabel("No of users");
plt.figure(figsize=(12,8));
```

```python
value_have = [i*(100/value_sum) for i in value_have]
```

```python
sns.barplot(x=value_have,y=key_have)
plt.title("Number of respondents using the languages")
plt.xlabel("Percent");
```

```python
value_want = []
key_want = []
value_sum2 = 0
for k,v in lang_dict_want.items():
    value_want.append(v)
    key_want.append(k)
    value_sum2 = value_sum + v
```

```python
sns.barplot(x=value_want,y=key_want)
plt.title("Number of respondents wanting to use the languages")
plt.xlabel("No of users");
```

```python
value_want = [i*(100/value_sum2) for i in value_want]
```

```python
sns.barplot(x=value_want,y=key_want)
plt.title("Number of respondents wanting to use the languages")
plt.xlabel("Percent");
```

```python
fig, axes = plt.subplots(1,2)
plt.tight_layout(pad=5)
sns.barplot(x=value_want,y=key_want, ax =axes[0])
axes[0].set_title("Who want to use in percent");


sns.barplot(x=value_have,y=key_have, ax=axes[1])
axes[1].set_title("who have used in percent");
```
